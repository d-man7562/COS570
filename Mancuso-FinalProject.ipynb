```python
import random

def nim_robot(gamelist):

    #Analyzes the game state and determines the robot's move.

    print("Nim Robot's turn.")
    # Simple winning move: if only one pile remains, take all of it.
    if len(gamelist) == 1:
        print(f"Nim Robot can win from this position. It is taking all {gamelist[0]} stones from pile 0.")
        gamelist[0] = 0
        return

    nim_sum = 0
    for value in gamelist:
        nim_sum ^= value
    print("XOR of every bit: ", nim_sum)

    if nim_sum == 0:
        print("I am in a P-position.")
        # If in a P-position, make a sub-optimal move to maintain game flow
        pile_index = -1
        while pile_index == -1 or gamelist[pile_index] == 0:
            pile_index = random.randint(0, len(gamelist) - 1)
        stones_to_remove = random.randint(1, gamelist[pile_index])

    else:
        print("I am in an N-position.")
        # Find the optimal move
        found_move = False
        for i in range(len(gamelist)):
          #target size is index within gamelist XORed by the Nim sum
            target_size = gamelist[i] ^ nim_sum
            #it will return a binary value that is less than our particular gamelist index
            if target_size < gamelist[i]:
              #we will subtract it
                stones_to_remove = gamelist[i] - target_size
                pile_index = i
                found_move = True
                break

        if not found_move:
            # Fallback to a sub-optimal move if for some reason an optimal move isn't found
            pile_index = random.randint(0, len(gamelist) - 1)
            stones_to_remove = random.randint(1, gamelist[pile_index])

    gamelist[pile_index] -= stones_to_remove
    if gamelist[pile_index] == 0:
      del gamelist[pile_index]
    print(f"Nim Robot takes {stones_to_remove} stones from pile {pile_index}. New gamelist: {gamelist}")


def nim_game(gamelist,turn= 0):
    #gamelist = [n] * piles
    turn = 0  # 0 for player, 1 for robot

    print("Gamelist:", gamelist)
    print("Make a move with a pile index (0 index) and amount of stones to remove (e.g., 0 2).")

    while sum(gamelist) > 0:
        print("-" * 30)

        # Player's turn
        if turn == 0:
            nim_sum = 0
            for value in gamelist:
              nim_sum ^= value
            print("XOR of every bit: ", nim_sum)

            player_move = input("Your turn. Please make a move: ")
            try:
                play_list = player_move.split()
                if len(play_list) != 2:
                    print("Invalid input. Please enter exactly two numbers.")
                    continue
                pile_index = int(play_list[0])
                stones_to_remove = int(play_list[1])

                if not (0 <= pile_index < len(gamelist)):
                    print("Invalid move: Pile index is out of range.")
                    continue
                if not (0 < stones_to_remove <= gamelist[pile_index]):
                    print("Invalid move: Not enough stones in the pile or invalid amount.")
                    continue

                gamelist[pile_index] -= stones_to_remove
                if gamelist[pile_index] == 0:
                    del gamelist[pile_index]
                turn = 1 # Switch to robot's turn

            except (ValueError, IndexError):
                print("Invalid input. Please enter two valid numbers.")

        # Robot's turn
        else:
            nim_robot(gamelist)
            if sum(gamelist) == 0:
                break
            turn = 0 # Switch to player's turn

    print("-" * 30)
    if turn == 0:
        print("Congratulations! You won!")
    else:
        print("I win! Better luck next time!")
```
