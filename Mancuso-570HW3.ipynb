```python
def find_stops(lst):
  ar = []
  for i, color in enumerate(lst):
    if color ==  "r":
      stop_r = ("r right stop: ",lst[i:])
      ar.append(stop_r)
      stop_l = ("r left stop: ",lst[:i])
      ar.append(stop_l)
    if color =="l":
      stop_r = ("l right stop: ",lst[i:])
      ar.append(stop_r)
      stop_l = ("l left stop: ",lst[:i])
      ar.append(stop_l)
  return ar
```

```python
from functools import lru_cache
import copy

def get_outcome_class_snort(board, left_to_move=True):
    rows, cols = len(board), len(board[0])
    key = (tuple(tuple(row) for row in board), left_to_move)
    if not hasattr(get_outcome_class_snort, "memo"):
        get_outcome_class_snort.memo = {}
    memo = get_outcome_class_snort.memo
    if key in memo:
        return memo[key]

    my_color = 1 if left_to_move else 2
    opp_color = 2 if left_to_move else 1

    found_move = False
    my_options = []
    opp_options = []

    for r in range(rows):
        for c in range(cols):
            if board[r][c] != 0:
                continue
            legal = True
            for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
                nr, nc = r+dr, c+dc
                if 0 <= nr < rows and 0 <= nc < cols and board[nr][nc] == opp_color:
                    legal = False
                    break
            if legal:
                found_move = True
                new_board = copy.deepcopy(board)
                new_board[r][c] = my_color
                result = get_outcome_class_snort(new_board, not left_to_move)
                if left_to_move:
                    my_options.append(result)
                else:
                    my_options.append(result)
    if not found_move:
        memo[key] = 'P'
        return 'P'

    if left_to_move:
        left_opts = my_options
        right_opts = []
        for r in range(rows):
            for c in range(cols):
                if board[r][c] != 0:
                    continue
                legal = True
                for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
                    nr, nc = r+dr, c+dc
                    if 0 <= nr < rows and 0 <= nc < cols and board[nr][nc] == 1:
                        legal = False
                        break
                if legal:
                    new_board = copy.deepcopy(board)
                    new_board[r][c] = 2
                    outcome = get_outcome_class_snort(new_board, True)
                    right_opts.append(outcome)
    else:
        right_opts = my_options
        left_opts = []
        for r in range(rows):
            for c in range(cols):
                if board[r][c] != 0:
                    continue
                legal = True
                for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
                    nr, nc = r+dr, c+dc
                    if 0 <= nr < rows and 0 <= nc < cols and board[nr][nc] == 2:
                        legal = False
                        break
                if legal:
                    new_board = copy.deepcopy(board)
                    new_board[r][c] = 1
                    outcome = get_outcome_class_snort(new_board, False)
                    left_opts.append(outcome)

    if all(opt in ['R', 'N'] for opt in left_opts) and any(opt in ['P', 'R'] for opt in right_opts):
        memo[key] = 'R'
        return 'R'
    if all(opt in ['L', 'N'] for opt in right_opts) and any(opt in ['P', 'L'] for opt in left_opts):
        memo[key] = 'L'
        return 'L'
    if 'P' in my_options:
        memo[key] = 'N'
        return 'N'
    memo[key] = 'P'
    return 'P'
```
