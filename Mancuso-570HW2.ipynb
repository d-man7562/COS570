```python
def find_period_and_length(lst):
    n = len(lst)
    if n == 0:
        return [],
    if n == 1:
        return lst, 1
    for p_len in range(1, n + 1):
        if n % p_len == 0:
            potential_period = lst[:p_len]
            is_repeating = True
            for i in range(p_len, n, p_len):
                if lst[i:i + p_len] != potential_period:
                    is_repeating = False
                    break
            if is_repeating:
                return potential_period, p_len



    return None, 0
```

```python
def find_preperiod_and_period(lst):
    n = len(lst)
    if n <= 1:
        return  False

    for pp_len in range(n):
        pre_period_candidate = lst[:pp_len]
        for p_len in range(1, (n - pp_len) // 2 + 1):
            if (n - pp_len) < 2 * p_len:
                continue
            potential_period = lst[pp_len : pp_len + p_len]

            is_periodic_match = True
            for i in range(pp_len + p_len, n, p_len):
                current_segment = lst[i : i + p_len]
                if current_segment != potential_period:
                    is_periodic_match = False
                    break
            if is_periodic_match:

                remaining_len = (n - pp_len) % p_len
                if remaining_len > 0:
                    if lst[n - remaining_len : n] != potential_period[:remaining_len]:
                        is_periodic_match = False

                if is_periodic_match:
                    return pre_period_candidate, potential_period, p_len, True
    return [], [], False
```
