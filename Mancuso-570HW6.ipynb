```python
import random as rand
import random
import sys
def create_graph(num):
  if num<20:
    return "cannot return graph less than size 20"
  #i am choosing max 100. i can choose sys.maxsize also
  #but it will complicate load time
  num_verts = random.randint(20,num)
  dicts ={}

  for i in range(num_verts):
  #from 0 to num_verts, initialze dict key and set values to empty list
    dicts[i] = []
  #for each key in the dictionary, we will
  #initailize a random amount of edges from minimum
  #2 up to num_verts number of edges
  #the reason minimum amount of edges is 2 is so that
  #every time, it guarantees that number of edges will be 2n
  #despite how unlikely it will be that it is not
  #if i happen to use range 1 to num_verts
  for key in dicts:
    #we need minimum 2 edges per vert so that it fits E=2V
    NUM_EDGES_PER_VERT = rand.randint(2,num_verts)
    for j in range(NUM_EDGES_PER_VERT):
      #connects to any edge within vert 0 to max vertex number
      connects_to = rand.randint(0,num_verts)
      #from 0 to j, we will assign random connection to key
      #with maximum connection being highest vert number
      dicts[key].append(connects_to)
  return dicts



```

```python
def dijkstra_like(dicts):
    visited_list = []
    current_node = 0
    current = dicts[current_node]
    total_weight = 0

    while True:
        visited_list.append(current_node)

        options = [k for k in current if k not in visited_list]
        if not options:
            print("Nowhere to go from here.")
            break

        # Choose the neighbor with the smallest edge weight
        next_node = min(options, key=lambda k: current[k])
        total_weight += current[next_node]
        current_node = next_node
        current = dicts[current_node]

    return total_weight, visited_list
```

```python
def dijkstra_path(graph, start, end):
    # distance from start to every node initially at infinity
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # track the previous node as path
    previous = {node: None for node in graph}

    # Keep track of unvisited nodes
    unvisited = set(graph.keys())

    while unvisited:
        # choose unvisited node with the smallest distance
        current = min(
            (node for node in unvisited if distances[node] != float('inf')),
            key=lambda node: distances[node],
            default=None
        )

        if current is None or current == end:
            break

        unvisited.remove(current)

        for neighbor, weight in graph[current].items():
            if neighbor in unvisited:
                new_dist = distances[current] + weight
                if new_dist < distances[neighbor]:
                    distances[neighbor] = new_dist
                    previous[neighbor] = current

    # reconstruct shortest path
    path = []
    curr = end
    while curr is not None:
        path.append(curr)
        curr = previous[curr]
#list order from path start to path end
    path = path[::-1]
# unreachable
    if distances[end] == float('inf'):
        return None, float('inf')

    return path, distances[end]
```

```python
import heapq

def reconstruct_path(previous, target):
    path = []
    while target is not None:
        path.append(target)
        target = previous[target]
    return path[::-1]

def dijkstras(graph, start):
    # Collect all nodes: top-level keys and all neighbors
    all_nodes = set(graph.keys())
    for edges in graph.values():
        all_nodes.update(edges.keys())

    # Init distance and previous node mappings
    distances = {node: float('inf') for node in all_nodes}
    previous = {node: None for node in all_nodes}
    distances[start] = 0

    # Min-heap: (distance, node)
    heap = [(0, start)]

    while heap:
        current_distance, current_node = heapq.heappop(heap)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph.get(current_node, {}).items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_node
                heapq.heappush(heap, (distance, neighbor))

    # Print all reachable paths
    print(f"Shortest paths from node {start}:\n")
    for node in sorted(all_nodes):
        if distances[node] == float('inf'):
            print(f"Node {node}: unreachable")
        else:
            path = reconstruct_path(previous, node)
            path_str = " → ".join(map(str, path))
            print(f"Node {node}: distance = {distances[node]}, path = {path_str}")

    return distances, previous
```
